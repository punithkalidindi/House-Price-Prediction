```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
np.random.seed(42)
num_samples = 1000
bedrooms = np.random.randint(1, 8, size=num_samples)
costs = 50000 + bedrooms * 50000
```

```python
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(bedrooms, costs, epochs=50)
```

```python
num_bedrooms_to_predict = np.array([7])
predicted_cost = model.predict(num_bedrooms_to_predict)
```

```python
for i in range(len(num_bedrooms_to_predict)):
    print(f"Actual cost for a {num_bedrooms_to_predict[i]} bedroom  house: ${costs[i]:.2f}")
    print(f"Predicted cost for a {num_bedrooms_to_predict[i]} bedroom  house: ${predicted_cost[i][0]:.2f}")
    print()
```
